```python
import pandas as pd
import torch
from transformers import BertTokenizer, BertForSequenceClassification, AdamW
from torch.utils.data import DataLoader, TensorDataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

labels_codes = {
    "Политика": 1,
    "Экономика": 2,
    "Общество": 3,
    "Закон и право": 4,
    "Кино": 5,
    "Телевидение": 6,
    "Персоны": 7,
    "События": 8,
    "Бренды": 9,
    "Наука": 10,
    "Гаджеты": 11,
    "Coцсети": 12,
    "Технологии": 13,
    "Опросы": 14,
    "Головоломки": 15,
    "Дом": 16,
    "Транспорт": 17,
    "Погода": 18,
    "Рецепты": 19,
    "Мода": 20,
    "Красота": 21,
    "Животные": 22
}

# Загрузка предварительно обученной модели RuBERT
# Загрузка предварительно обученной модели RuBERT
tokenizer = BertTokenizer.from_pretrained("DeepPavlov/rubert-base-cased")
model = BertForSequenceClassification.from_pretrained("DeepPavlov/rubert-base-cased", num_labels=22)  # Установка количества классов

# Загрузка вашего датасета

data1 = pd.read_csv("summary-1_10000.csv")
data2 = pd.read_csv("summary-2.csv")
data = pd.concat([data1, data2])
print(data.shape)
data = data.dropna()
print(data.shape)

# Предобработка данных
texts = data['summary'].tolist()  # Используйте соответствующий столбец для текстов
labels = data['meta_category'].apply(lambda x: labels_codes[x]).tolist()  # Используйте соответствующий столбец для меток

# Разбиение на обучающую и тестовую выборки
train_texts, test_texts, train_labels, test_labels = train_test_split(texts, labels, test_size=0.2, random_state=42)

# Токенизация текстов с обрезанием до максимальной длины
max_length = 32
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=max_length)
test_encodings = tokenizer(test_texts, truncation=True, padding=True, max_length=max_length)

# Преобразование меток в индексы классов
label_map = {label: i for i, label in enumerate(set(labels))}
train_labels = [label_map[label] for label in train_labels]
test_labels = [label_map[label] for label in test_labels]

# Преобразование меток в тензоры
train_labels = torch.tensor(train_labels)
test_labels = torch.tensor(test_labels)

# Создание датасетов
train_dataset = TensorDataset(torch.tensor(train_encodings['input_ids']),
                              torch.tensor(train_encodings['attention_mask']), train_labels)
test_dataset = TensorDataset(torch.tensor(test_encodings['input_ids']),
                             torch.tensor(test_encodings['attention_mask']), test_labels)

# Создание DataLoader'ов
batch_size = 8
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Параметры обучения
learning_rate = 2e-5
num_epochs = 3

# Оптимизатор и функция потерь
optimizer = AdamW(model.parameters(), lr=learning_rate)
loss_fn = torch.nn.CrossEntropyLoss()

# Обучение модели
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
print(device)

for epoch in range(num_epochs):
    model.train()
    print("Epocha:", epoch)
    for batch in train_loader:
        input_ids, attention_mask, labels = [tensor.to(device) for tensor in batch]

        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

# Оценка производительности модели
model.eval()
predictions = []
true_labels = []

with torch.no_grad():
    for batch in test_loader:
        input_ids, attention_mask, labels = [tensor.to(device) for tensor in batch]
        outputs = model(input_ids, attention_mask=attention_mask)
        _, predicted = torch.max(outputs.logits, 1)
        predictions.extend(predicted.cpu().numpy())
        true_labels.extend(labels.cpu().numpy())

# Оценка производительности модели
print("Accuracy:", accuracy_score(true_labels, predictions))
print("\nClassification Report:")
print(classification_report(true_labels, predictions))
```
